### Quebrando a entrada em frames

```python
frame_length = 2048

dataset = []
for sample in x:
  length = sample.shape[0]
  num_frames = int(length / frame_length)
  amostras_uteis = num_frames * frame_length
  sample_cutted = sample[:amostras_uteis]
  sample_cutted = sample_cutted.reshape(num_frames, frame_length)
  dataset.append(sample_cutted)

x = np.array(dataset)
print(x.shape)
x[0:5]
```

## Construindo o modelo

```python
from keras.layers import TimeDistributed, Dense, Input, Lambda, BatchNormalization, Concatenate, Conv1D, ZeroPadding1D
from keras.models import Model
from tensorflow.keras.utils import plot_model
from keras.callbacks import EarlyStopping
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Dense, Input, Lambda, TimeDistributed
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model

activation_function = "relu"

input_layer = Input(shape=(x.shape[1], x.shape[2]))

input_layer_extrator = Input(shape=[x.shape[2]])
x_0 = Dense(int(x.shape[2]/10), activation='relu', name='encoder', use_bias=True)(input_layer_extrator)
extrator = Model(input_layer_extrator, x_0, name='extrator')

features_padrao = TimeDistributed(extrator, name="features_padrao")(input_layer)
media_features_padrao = Lambda(tf.reduce_mean)(features_padrao)

# saida = Dense(x.shape[1], activation='relu', name='encoder', use_bias=True)(media_features_padrao)

# model = Model(input_layer, saida)

# # Compilando o modelo
# model.compile(loss='mse', optimizer='Adam', metrics=['mse', 'mae', 'rmse'])
# plot_model(model, show_shapes=True, expand_nested=True)
```

```python
def extrator_features(input_dims, activation):
  input_layer = Input(shape=[input_dims])

  x_0 = Dense(input_dims, activation=activation, use_bias=True)(input_layer)
  x_bottleneck = Dense(input_dims, activation=activation, name='encoder', use_bias=True)(x_0)

  return Model(input_layer, x_bottleneck, name="features")

def regressor(input_dims, output_dims, activation):
  input_layer = Input(shape=[input_dims])

  x_0 = Dense(int(input_dims), activation=activation, use_bias=True)(input_layer)
  x_2 = Dense(int(input_dims*2), activation=activation, use_bias=True)(x_0)
  x_3 = Dense(int(input_dims), activation=activation, use_bias=True)(x_2)
  x_4 = Dense(int(input_dims/2), activation=activation, use_bias=True)(x_3)
  x_5 = Dense(int(input_dims/4), activation=activation, use_bias=True)(x_4)
  x_6 = Dense(int(input_dims/8), activation=activation, use_bias=True)(x_5)
  saidas = Dense(output_dims, activation=activation, name='regressor_saidas', use_bias=True)(x_6)

  return Model(input_layer, saidas, name="regressor")

def my_mean(x):
  return tf.reduce_mean(x, axis=1)

def build_models(input_len, input_dims, output_dims, activation):
  # Camadas de entrada
  input_layer = Input(shape=(input_len, input_dims))

  x_n = BatchNormalization()(input_layer)

  # Kernels
  extrator = extrator_features(input_dims, activation)

  # Features
  features = TimeDistributed(extrator, name="features_padrao")(x_n)

  # Juntando as features
  media_features = Lambda(my_mean)(features)

  # Regrssão
  regressao = regressor(media_features.shape[1], output_dims, activation)

  saida = regressao(media_features)
  
  return (
      Model(input_layer, saida, name="regressao"),
      Model(input_layer, media_features, name="projecao")
  )

model, features = build_models(x.shape[1], x.shape[2], y.shape[1], 'elu')
model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse', 'mean_absolute_percentage_error'])
plot_model(model, show_shapes=True, expand_nested=True)
```

```python
# Callback para recuperar o melhor peso, e parar quando ficar três épocas sem melhora
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)

# Treinando o modelo
history = model.fit(x, y, epochs=50, validation_split = 0.2, callbacks=[callback])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```
