# Testando a implementação manual da síntese FM

## Imports

```python
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt

from fm_synth import FMSynth, SAMPLE_RATE
```

## Analisando uma senóide pura (que serve para melhor compreenção do efeito da FM)

### Testando a síntese por meio da símples audição

```python
fm_synth = FMSynth(
    amplitude1=0,
    frequency1=0,
    beta2=0.0,
    amplitude2=0,
    frequency2=0,
    beta3=0,
    amplitude3=0,
    frequency3=0,
    beta_carrier=0,
    amplitude_carrier=0.8,
    attack=0.01,
    decay=0.2,
    sustain=0.2,
    release=0.3,
)
signal = fm_synth.synth_alg1(3, 440.0)

ipd.Audio(signal, rate=SAMPLE_RATE)
```

### Analisando o spectro do sinal (FFT)

#### Rodando o FFT

```python
qtd_amostras = len(signal)  # Número de amostras
domain_frequency_signal = np.fft.fft(signal)  # Aplicar a FFT
frequencies = np.fft.fftfreq(qtd_amostras, 1/SAMPLE_RATE)  # Obter as frequências correspondentes
```

#### Plotando o gráfico do espectro de frequencias

```python
# Apenas a metade positiva importa (frequências reais); mas vamos plotar menos do que isso (para melhorar o zoom do gráfico)
plt.figure(figsize=(12, 6))
# plt.plot(frequencies[:qtd_amostras//2], np.abs(domain_frequency_signal[:qtd_amostras//2]))  
plt.plot(frequencies[:qtd_amostras//8], np.abs(domain_frequency_signal[:qtd_amostras//8]))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.title('Espectro de Frequências (FFT)')
plt.grid(True)
plt.show()
```
